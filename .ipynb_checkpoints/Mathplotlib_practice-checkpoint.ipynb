```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create a Line Chart Using the Object-Oriented Interface

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create a Line Chart Using the Object-Oriented Interface

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="*", color="green", label= "Boston")
ax.set_xlabel("Date")
ax.set_ylabel(" Fares ($)")
ax.set_title("PyBer Fare Month")
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="purple", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer by Month")
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, label="Chicago", color="cyan")
ax.invert_yaxis()
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare Month")
```

```python
plt.scatter(x_axis, y_axis, c="r", label="Chicago", s= [i*3 for i in y_axis])
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="blue", label="Chicago", s=[i*5 for i in y_axis], edgecolor="black")
ax.set_xlabel("Fares ($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare Month")
ax.invert_yaxis()
```
